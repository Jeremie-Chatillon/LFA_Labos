## Introduction

In this lab we will look at the interpretability aspect of the fuzzy systems. Indeed, having a FIS that describes accurately a problem is good but having a FIS that is built with interpretability in mind is better.

As you have seen in the lectures, the fuzzy systems are grey-box models meaning that you can "read" them as a human and try to understand how they perform their decisions. As a human you cannot understand a too complex fuzzy system (e.g. too much rules, too much variables per rule, too much linguistic labels per linguistic variables,...).

The main idea here is to add constraints on how to build a fuzzy system. Those constraints will lead to a better interpretability with a possible loss of accuracy i.e. the "famous" **performance-interpretability trade-off**

````python

# Visualisation des règles fouries
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='3d'))
show_surface(fis, ax=ax, title="Car Problem Consequent", x_label="temperature", y_label="sunshine", z_label="tourists",
                 n_pts=15, x_range=(17, 29), y_range=(30, 100), z_range=(0, 100))
````

Nous pouvons constater qu'une large partie des valeurs ne sont pas représentées.
Nous avons décidé de mettre une règle par défaut. <br>
La principale raison est qu'il est plus simple d'écrire une règle qui prend tous les cas non traîté que plusieurs règles pour chacun des cas.<br>
De plus, nous ne sommes pas des experts dans le "tourisme", donc nous nous sommes dit que les règles données étaient déjà les cas spéciaux/extrèmes. Donc il manque plus qu'à fixer une règle qui dicte les valeurs qui représentent "tourism = medium".

Nous avons fait cette partie après la 5.1 pour pouvoir appliquer la règle par défaut.

````python
from fuzzy_systems.core.rules.default_fuzzy_rule import DefaultFuzzyRule
from fuzzy_systems.core.fis.singleton_fis import SingletonFIS
from fuzzy_systems.core.membership_functions.singleton_mf import SingletonMF
````

````python
# Cariable ouput en Singlton
lv_tourists = LinguisticVariable(name="tourists", ling_values_dict={
    "low": SingletonMF(0),
    "medium": SingletonMF(50),
    "high": SingletonMF(100)
})

# Recopie des règles fournies
r1 = FuzzyRule(
    ants=[
        Antecedent(lv_temperature, "hot"),
        Antecedent(lv_sunshine, "sunny")
    ],
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_tourists, "high")
    ],
    impl_func=MIN
)

r2 = FuzzyRule(
    ants=[
        Antecedent(lv_temperature, "cold"),
        Antecedent(lv_sunshine, "cloudy")
    ],
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_tourists, "low"),
    ],
    impl_func=MIN
)

# Création de la règle par défaut: Tout ce qui est pas dit = medium
dfRule = DefaultFuzzyRule(cons=[Consequent(lv_tourists, 'medium')], impl_func=MIN)

# Création du FIS
fisSinglDf = SingletonFIS(rules=[r1, r2], default_rule = dfRule) 
````

````python

# Affichage du nouveau FIS
fig2, ax2 = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='3d'))
show_surface(fisSinglDf, ax=ax2, title="FIS default Rules", x_label="temperature", y_label="sunshine", z_label="tourists",
                 n_pts=15, x_range=(17, 29), y_range=(30, 100), z_range=(0, 100))
````

Nous pouvons constater que dorénavent. Toutes les valeurs de "tourist" sont représentées => OK
