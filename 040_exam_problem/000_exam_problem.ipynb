#### Linguistic variables
difficulty: {easy, medium, hard, very_hard} ∈ [0;10] <br>
importance: {null, medium, high} ∈ [0;10] <br>
remaining_work: {nothing, few, bit, more, lot} ∈ [0;100] <br>
time_to_exam: {soon, middle , futur) ∈ [0;30] <br>
#### output
Priority: {low, medium, high} <br>

#### Rules:
> rule 1: IF time_to_exam is soon THEN Priority is high

> rule 2: IF time_to_exam is futur THEN Priority is low

> rule 3: IF remaining_work is lot OR difficulty is very_hard THEN Priority is high

> rule 4: IF importance is null OR remaining_work is nothing THEN Priority is low

> rule 5 IF importance is medium AND remaining_work is few AND difficulty is medium THEN Priority is medium

Grace à ces règles, nous avons définit que **time_to_exam** est la variable la plus importante. Car dans la vie, on peut tout réaliser (ou prèsque) si on a le temps. L'importance à peu d'importance. 

````python
# Step 1: define the linguistic variables
lv_difficulty = LinguisticVariable(name="difficulty", ling_values_dict={
        "easy": LinPWMF([0, 1], [2, 0]),
        "medium": TrapMF(0, 2, 4, 6,),
        "hard": TrapMF(4, 6, 8, 10),
        "very_hard": LinPWMF([8, 0], [10, 1]),
    })

lv_importance = LinguisticVariable(name="importance", ling_values_dict={
        "null": LinPWMF([0, 1], [2, 0]),
        "medium": TrapMF(0, 2, 8, 10,),
        "high": LinPWMF([8, 0],[10,1]),
    })
lv_remaining_work = LinguisticVariable(name="remaining_work", ling_values_dict={
        "nothing": LinPWMF([0, 1], [5, 0]),
        "few": LinPWMF([0, 0], [5, 1], [20, 0]),
        "bit": LinPWMF([5, 0], [20, 1],[40,0]),
        "more": LinPWMF([20,0],[40,1],[60,0]),
        "lot": LinPWMF([40, 0], [100, 1])
    })

lv_time_to_exam = LinguisticVariable(name="time_to_exam", ling_values_dict={
        "soon": LinPWMF([0, 1], [6, 0]),
        "middle": LinPWMF([0, 0], [6, 1], [30, 0]),
        "futur": LinPWMF([6, 0],[30,1]),
    })

lv_priority = ThreePointsLV("Priority", 0.0, 0.5, 1.10)

fig = plt.figure(figsize=(15.0, 5.0))
fig2 = plt.figure(figsize=(15.0, 5.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)
axes4 = fig2.add_subplot(1, 3, 1)
axes5 = fig2.add_subplot(1, 3, 2)

LinguisticVariableViewer(lv_difficulty, axes1)
LinguisticVariableViewer(lv_importance, axes2)
LinguisticVariableViewer(lv_remaining_work, axes3)
LinguisticVariableViewer(lv_time_to_exam, axes4)
LinguisticVariableViewer(lv_priority, axes5)
````

````python
# Règles finales, ce sont celles qui correspondent le mieux à notre expertise

# rule 1: IF time_to_exam is soon THEN Priority is high
# Define the rules
r1 = FuzzyRule(
    ants=[
        Antecedent(lv_time_to_exam, "soon"),
    ],
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_priority, "high"),
    ],
    # implication function or "how the activated antecedents value will be combined with the consequent(s)
    impl_func=MIN
)
# rule 2: IF time_to_exam is futur THEN Priority is low
r2 = FuzzyRule(
    ants=[
        Antecedent(lv_time_to_exam, "futur"),
    ],
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_priority, "low"),
    ],
    # implication function or "how the activated antecedents value will be combined with the consequent(s)
    impl_func=MIN
)

    
#rule 3: IF remaining_work is lot OR difficulty is very_hard THEN Priority is high
r3 = FuzzyRule(
    ants=[
        Antecedent(lv_remaining_work, "lot"),
        Antecedent(lv_difficulty, "very_hard"),
    ],
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_priority, "high"),
    ],
    # implication function or "how the activated antecedents value will be combined with the consequent(s)
    impl_func=MIN
)

#rule 4: IF importance is null OR remaining_work is nothing THEN Priority is low
r4 = FuzzyRule(
    ants=[
        Antecedent(lv_importance, "null"),
        Antecedent(lv_remaining_work, "nothing"),
    ],
    ant_act_func=OR_max,
    cons=[
        Consequent(lv_priority, "low"),
    ],
    # implication function or "how the activated antecedents value will be combined with the consequent(s)
    impl_func=MIN
)

#rule 5 IF importance is medium OR remaining_work is few OR difficulty is medium THEN Priority is medium
r5 = FuzzyRule(
    ants=[
        Antecedent(lv_importance, "medium"),
        Antecedent(lv_remaining_work, "few"),
        Antecedent(lv_difficulty, "medium"),
    ],
    ant_act_func=AND_min,
    cons=[
        Consequent(lv_priority, "medium"),
    ],
    # implication function or "how the activated antecedents value will be combined with the consequent(s)
    impl_func=MIN
)

fis = FIS(
    rules=[r1, r2, r3, r4, r5],
    aggr_func=np.max, # aggregation function or "how the implicated consequent(s) will be combined together"
    defuzz_func=COA_func # defuzzification or "how the "final" membership function created after the aggregation step will be interpreted as a output crisp value
)
````

````python
def input_var(diff, imp, remaining, time):
    return {'difficulty': diff, 'importance': imp, 'remaining_work': remaining, 'time_to_exam': time}
       
````
